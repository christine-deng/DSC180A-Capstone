#### 1.1 Use Case

**In order to demonstrate how AIF360 can be used to detect and mitigate bias in classfier models, we adopt the following use case:**

* Data scientist develops a 'fair' healthcare utilization scoring model with respect to defined protected classes. Fairness may be dictated by legal or government regulations, such as a requirement that additional care decisions be not predicated on factors such as race of the patient.
* Developer takes the model AND performance characteristics / specs of the model (e.g. accuracy, fairness tests, etc. basically the model factsheet) and deploys the model in an enterprise app that prioritizes cases for care management.
* The app is put into production and starts scoring people and making recommendations. 
* Explanations are generated for each recommendation
* Both recommendations and associated explanations are given to nurses as a part of the care management process. The nurses can evaluate the recommendations for quality and correctness and provide feedback.
* Nurse feedback as well as analysis of usage data with respect to specs of the model w.r.t accuracy and fairness is communicated to AI Ops specialist and LOB user periodically.
* When significant drift in model specs relative to the model factsheet is observed, the model is sent back for retraining.

#### 1.3 Methodology 

For each dataset, the sensitive attribute is 'RACE' constructed as follows: 'Whites' (privileged class) defined by the features RACEV2X = 1 (White) and HISPANX = 2 (non Hispanic); 'Non-Whites' that included everyone else.  

* Along with race as the sensitive feature, other features used for modeling include demographics  (such as age, gender, active duty status), physical/mental health assessments, diagnosis codes (such as history of diagnosis of cancer, or diabetes), and limitations (such as cognitive or hearing or vision limitation).
* To measure utilization, a composite feature, 'UTILIZATION', was created to measure the total number of trips requiring some sort of medical care by summing up the following features: OBTOTV15(16), the number of office based visits;  OPTOTV15(16), the number of outpatient visits; ERTOT15(16), the number of ER visits;  IPNGTD15(16), the number of inpatient nights, and  + HHTOTD16, the number of home health visits.
* The model classification task is to predict whether a person would have 'high' utilization (defined as UTILIZATION >= 10, roughly the average utilization for the considered population). High utilization respondents constituted around 17% of each dataset.
* To simulate the scenario, each dataset is split into 3 parts: a train, a validation, and a test/deployment part.

**We assume that the model is initially built and tuned using the 2015 Panel 19 train/test data**
* It is then put into practice and used to score people to identify potential candidates for care management. 
* Initial deployment is simulated to 2015 Panel 20 deployment data. 
* To show change in performance and/or fairness over time, the 2016 Panel 21 deployment data is used. 
* Finally, if drift is observed, the 2015 train/validation data is used to learn a new model and evaluated again on the 2016 deployment data

```python
default_mappings = {
    'label_maps': [{1.0: '>= 10 Visits', 0.0: '< 10 Visits'}],
    'protected_attribute_maps': [{1.0: 'White', 0.0: 'Non-White'}]}

def default_preprocessing19(df):
    """
    1.Create a new column, RACE that is 'White' if RACEV2X = 1 and HISPANX = 2 i.e. non Hispanic White
      and 'non-White' otherwise
    2. Restrict to Panel 19
    3. RENAME all columns that are PANEL/ROUND SPECIFIC
    4. Drop rows based on certain values of individual features that correspond to missing/unknown - generally < -1
    5. Compute UTILIZATION, binarize it to 0 (< 10) and 1 (>= 10)
    """
    def race(row):
        if ((row['HISPANX'] == 2) and (row['RACEV2X'] == 1)): 
            return 'White'
        return 'Non-White'

    df['RACEV2X'] = df.apply(lambda row: race(row), axis=1)
    df = df.rename(columns = {'RACEV2X' : 'RACE'})
    
    df = df[df['PANEL'] == 19]

    # RENAME COLUMNS
    df = df.rename(columns = {'FTSTU53X' : 'FTSTU', 'ACTDTY53' : 'ACTDTY', 'HONRDC53' : 'HONRDC', 'RTHLTH53' : 'RTHLTH',
                              'MNHLTH53' : 'MNHLTH', 'CHBRON53' : 'CHBRON', 'JTPAIN53' : 'JTPAIN', 'PREGNT53' : 'PREGNT',
                              'WLKLIM53' : 'WLKLIM', 'ACTLIM53' : 'ACTLIM', 'SOCLIM53' : 'SOCLIM', 'COGLIM53' : 'COGLIM',
                              'EMPST53' : 'EMPST', 'REGION53' : 'REGION', 'MARRY53X' : 'MARRY', 'AGE53X' : 'AGE',
                              'POVCAT15' : 'POVCAT', 'INSCOV15' : 'INSCOV'})

    df = df[df['REGION'] >= 0] # remove values -1
    df = df[df['AGE'] >= 0] # remove values -1

    df = df[df['MARRY'] >= 0] # remove values -1, -7, -8, -9

    df = df[df['ASTHDX'] >= 0] # remove values -1, -7, -8, -9

    df = df[(df[['FTSTU','ACTDTY','HONRDC','RTHLTH','MNHLTH','HIBPDX','CHDDX','ANGIDX','EDUCYR','HIDEG',
                             'MIDX','OHRTDX','STRKDX','EMPHDX','CHBRON','CHOLDX','CANCERDX','DIABDX',
                             'JTPAIN','ARTHDX','ARTHTYPE','ASTHDX','ADHDADDX','PREGNT','WLKLIM',
                             'ACTLIM','SOCLIM','COGLIM','DFHEAR42','DFSEE42','ADSMOK42',
                             'PHQ242','EMPST','POVCAT','INSCOV']] >= -1).all(1)]  #for all other categorical features, remove values < -1

    def utilization(row):
        return row['OBTOTV15'] + row['OPTOTV15'] + row['ERTOT15'] + row['IPNGTD15'] + row['HHTOTD15']

    df['TOTEXP15'] = df.apply(lambda row: utilization(row), axis=1)
    lessE = df['TOTEXP15'] < 10.0
    df.loc[lessE,'TOTEXP15'] = 0.0
    moreE = df['TOTEXP15'] >= 10.0
    df.loc[moreE,'TOTEXP15'] = 1.0

    df = df.rename(columns = {'TOTEXP15' : 'UTILIZATION'})
    return df
```

#### 2.3 Outlier Detection and Handling

Most categorical features in the unprocessed dataset have a variety of error codes, signified by different negative numbers, that signal different varieties of missingness. However, in this case, preprocessing has already occurred, unifying many of these negative numbers into a single category. We shall perform investigation to see if this is the case across all columns, or if instead there are some ununified negative values in the dataset.

```python
# Count the instances per column where there's a numeric value less than -1
df_panel_19_reduced.map(lambda x: x < -1 if not isinstance(x, str) else False).sum(), df_panel_20_reduced.map(lambda x: x < -1 if not isinstance(x, str) else False).sum()
```

We can see that there are three columns remaining that feature values less than -1. Let's see what these values are.

```python
negative_columns = df_panel_19_reduced[['PCS42', 'MCS42', 'K6SUM42']]
negative_columns_20 = df_panel_20_reduced[['PCS42', 'MCS42', 'K6SUM42']]
# Values below -1 within these columns
negative_columns[negative_columns < -1].value_counts()
```

```python
negative_columns_20[negative_columns_20 < -1].value_counts()
```

From our prior results, we know that none of the three of these features are categorical. As such, the negative values within them will not be properly processed as nulls, so we must swap these for NaNs to ensure valid predictions. However, before doing this, let's investigate all non-categorical features for negative values.

```python
# Excluding RACE since it's a string value
ordinal_features = df_panel_19_reduced.drop(categorical_features + ['RACE'], axis=1)
ordinal_features_20 = df_panel_20_reduced.drop(categorical_features + ['RACE'], axis=1)
for column in ordinal_features.columns:
    print(ordinal_features[ordinal_features < 0][column].value_counts())
```

```python
for column in ordinal_features.columns:
    print(ordinal_features_20[ordinal_features_20 < 0][column].value_counts())
```

Now we can substitute these for NaNs.

```python
for column in negative_columns.columns:
    df_panel_19_reduced[column] = df_panel_19_reduced[column].replace([-1.0, -9.0], np.nan)
    df_panel_20_reduced[column] = df_panel_20_reduced[column].replace([-1.0, -9.0], np.nan)
```

Now that we've made sure all the negative numbers are handled, we will check for NaNs outside of the columns where we've added them.

```python
df_panel_19_reduced.isna().sum()
```

```python
df_panel_20_reduced.isna().sum()
```

There are no NaNs to account for besides the ones we have intentionally added.

To detect outliers in the dataset, we will begin by investigating the frequency of how often each category comes up for the categorical features. We will exclude columns with no categories below a conservative frequency cutoff of 10%, so as to err on the side of investigating more features than not.

```python
for column in categorical_features:
    frequencies = df_panel_19_reduced[column].value_counts(dropna=False, normalize=True)
    # If there is a low frequency category
    if sum(frequencies < 0.1) > 0:
        print(df_panel_19_reduced[column].value_counts(dropna=False, normalize=True))
```

```python
for column in categorical_features:
    frequencies = df_panel_20_reduced[column].value_counts(dropna=False, normalize=True)
    # If there is a low frequency category
    if sum(frequencies < 0.1) > 0:
        print(df_panel_20_reduced[column].value_counts(dropna=False, normalize=True))
```

The MARRY column stands out due to its myriad of outliers. [Upon consulting the data file](https://meps.ahrq.gov/mepsweb/data_stats/download_data_files_codebook.jsp?PUFId=H181&varName=MARRY15X), it turns out that values 7, 8, 9, and 10 (Married in Round, Widowed in Round, Divorced in Round, Separated in Round) signify that the person's relationship status changed during the survey to what is specified in the value's name. Since these instances are so few, though, it is possible that treating them as their own distinct categories could lead to overfitting in our results. As such, we will unify each with its non-in round counterpart.

```python
for value in [7, 8, 9, 10]:
    # For each of these, its corresponding non-round value is itself subtracted by 6
    df_panel_19_reduced['MARRY'] = df_panel_19_reduced['MARRY'].replace(value, value - 6)
    df_panel_20_reduced['MARRY'] = df_panel_20_reduced['MARRY'].replace(value, value - 6)
```

For FTSTU, no correction is needed, since its outliers are simply a result of [not much of the population being full-time students.](https://meps.ahrq.gov/mepsweb/data_stats/download_data_files_codebook.jsp?PUFId=H181&varName=FTSTU15X) The same as well for the outlier in ACTDTY, which is caused by [most of the population not being active-duty military personnel.](https://meps.ahrq.gov/mepsweb/data_stats/download_data_files_codebook.jsp?PUFId=H181&varName=ACTDTY31) Since the outlier here is incredibly tiny, though, consideration should be given as to whether it could lead to overfitting when predicting on that category in a model. The outlier in HONRDC [is caused for the same reason as in ACTDTY](https://meps.ahrq.gov/mepsweb/data_stats/download_data_files_codebook.jsp?PUFId=H181&varName=HONRDC31) and merits the same consideration. 

Substantive outliers exist in RTHLTH, though two out of three [are caused by some results in the health survey being less popular than others.](https://meps.ahrq.gov/mepsweb/data_stats/download_data_files_codebook.jsp?PUFId=H181&varName=RTHLTH31) The only egregious outlier belongs to that of the nulls. MNHLTH is much the same, [just for perceived mental health instead.](https://meps.ahrq.gov/mepsweb/data_stats/download_data_files_codebook.jsp?PUFId=H181&varName=MNHLTH31) Since these two variables are ordinal, we will replace the nulls with NaNs so that it is clear that they have no ordinal relation to the other categories.

```python
df_panel_19_reduced['RTHLTH'] = df_panel_19_reduced['RTHLTH'].replace(-1, np.nan)
df_panel_19_reduced['MNHLTH'] = df_panel_19_reduced['MNHLTH'].replace(-1, np.nan)
df_panel_20_reduced['RTHLTH'] = df_panel_20_reduced['RTHLTH'].replace(-1, np.nan)
df_panel_20_reduced['MNHLTH'] = df_panel_20_reduced['MNHLTH'].replace(-1, np.nan)
```

From there, all categories up until PREGNT have outliers due to them [marking the existence of specific physical or mental health conditions in the person surveyed](https://meps.ahrq.gov/mepsweb/data_stats/download_data_files_codebook.jsp?PUFId=H181), conditions which the majority of the population does not have. As such, no processing needs to be done there. PREGNT itself also has an outlier because [most of the population is not pregnant](https://meps.ahrq.gov/mepsweb/data_stats/download_data_files_codebook.jsp?PUFId=H181&varName=PREGNT31). No processing needs to be done for this feature either.

WLKLIM, ACTLIM, SOCLIM, and COGLIM specify whether [the person surveyed has certain limitations atypical of the population at large](https://meps.ahrq.gov/mepsweb/data_stats/download_data_files_codebook.jsp?PUFId=H181), which naturally are going to be relatively uncommon within those surveyed. DFHEAR42 and DFSEE42 specify [serious sensory impairments](https://meps.ahrq.gov/mepsweb/data_stats/download_data_files_codebook.jsp?PUFId=H181&varName=DFHEAR42), which, as is to be expected, are only present within a small portion of the population. ADSMOK42's outliers are a result of it [signifying whether the surveyed person smokes,](https://meps.ahrq.gov/mepsweb/data_stats/download_data_files_codebook.jsp?PUFId=H181&varName=ADSMOK42) and most of the population does not. These also have outliers wherein the surveyed person did not answer or the question wasn't applicable to them, but imputing that value would bias the dataset. As we're dealing with categorical data, the only reasonable method of imputation would be the most common category, which is not wholly representative of the non-answerers. It is also possible that not answering the question could correlate with high or low impact.

EMPST signifies [whether the person in question was employed at the time of the survey.](https://meps.ahrq.gov/mepsweb/data_stats/download_data_files_codebook.jsp?PUFId=H181&varName=EMPST31) Its two outliers signify if the person in question found a job during the survey period, or if they had a job to return to during the survey period. Due to the extremely small size of these two outliers, it is likely they will lead to biased results if left in the dataset. To solve this problem, we will unify both of them with the "Employed" category.

```python
df_panel_19_reduced['EMPST'] = df_panel_19_reduced['EMPST'].replace(3, 1)
df_panel_19_reduced['EMPST'] = df_panel_19_reduced['EMPST'].replace(4, 1)
df_panel_20_reduced['EMPST'] = df_panel_20_reduced['EMPST'].replace(3, 1)
df_panel_20_reduced['EMPST'] = df_panel_20_reduced['EMPST'].replace(4, 1)
```

Finally, POVCAT [signifies a surveyed person's family's income bracket](https://meps.ahrq.gov/mepsweb/data_stats/download_data_files_codebook.jsp?PUFId=H181&varName=POVCAT15), with its outlier signifying if someone's family is below the poverty line. This outlier is to be expected, as most of the population is not in poverty, and it is large enough to the point where we can be relatively sure it will not bias the results.

### [7.](#Table-of-Contents) Explainability

In our research, achieving model explainability is paramount to ensure transparency and understanding of the bias mitigation techniques applied to logistic regression and random forest classifiers. Explainability allows stakeholders and end-users to comprehend the decision-making process, fostering trust and aiding in the identification and resolution of potential biases.

The bias mitigation technique employed in both logistic regression and random forest models during pre-processing is reweighing. Implemented at the pre-processing stage, reweighing involves assigning different weights to examples in each (group, label) combination to ensure fairness before classification. This technique specifically targets disparate impact, which measures the ratio of favorable outcomes for the unprivileged group to that of the privileged group.
